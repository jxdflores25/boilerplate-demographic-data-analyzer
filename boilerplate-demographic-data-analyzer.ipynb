```python
import pandas as pd
```

```python
df = pd.read_csv("adult.data.csv")
df.head()
```

**How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.**

```python
race_count = pd.Series(df['salary'].value_counts())
race_count
```

**¿Cuál es la edad promedio de los hombres?**

```python
average_age_men=df.loc[df['sex']=='Male','age'].mean().round(1)
average_age_men
```

**¿Cuál es el porcentaje de personas que tienen un grado de licencia?**

```python
percentage_bachelors=(df.loc[df['education']=='Bachelors','education'].count()/len(df)*100).round(1)
percentage_bachelors
```

**¿Qué porcentaje de personas con una educación avanzada (`Bachelors`, `Masters`, or `Doctorate` ) ganan más de 50k?**

**¿Qué porcentaje de personas sin una educación avanzada generan más de 50k?**

```python
education_person=df['education'].value_counts()

person_more_50K=df[df['salary'] == ">50K"]['education'].value_counts()
```

```python
#Hallar la cantidad total de personas con y sin educacion avanzada

person_of_higher_education=education_person[['Bachelors','Masters','Doctorate']].sum()
person_of_lower_education=education_person.sum() - person_of_higher_education

#Hallar la cantidad total de personas con y sin educacion avanzada que ganan mas de 50K

higher_education=person_more_50K[['Bachelors','Masters','Doctorate']].sum()
lower_education=person_more_50K.sum() - higher_education

#Hallar el porcentaje de personas con y sin educacion avanzada que ganan mas de 50K

higher_education_rich = (higher_education/person_of_higher_education * 100).round(1)
lower_education_rich = (lower_education/person_of_lower_education * 100).round(1)


print(higher_education_rich,lower_education_rich)
```

**¿Cuál es el mínimo número de horas que una persona trabaja por semana?**

```python
min_work_hours= df['hours-per-week'].min()
min_work_hours
```

**¿Qué porcentaje de personas que trabajan el mínimo de horas por semana tiene un salario de más de 50k?**

```python
num_min_workers= df[df['hours-per-week'] == 1]['hours-per-week'].count()

workers_more_50k = df[df['salary'] == '>50K' ]['hours-per-week'].value_counts()


rich_percentage = (workers_more_50k[1]/num_min_workers*100).round(1)
rich_percentage
```

**¿Qué país tiene el porcentaje más alto de personas que ganan >50k y cuál es ese porcentaje?**

```python
quantity_countries = df['native-country'].value_counts()
quantity_countries_more_50K = df[df['salary'] == '>50K']['native-country'].value_counts()

countries_percentage= (quantity_countries_more_50K/quantity_countries)*100

highest_earning_country = countries_percentage[ countries_percentage == countries_percentage.max()].index[0]
highest_earning_country_percentage = countries_percentage.max().round(1)

print(highest_earning_country,highest_earning_country_percentage)
```
